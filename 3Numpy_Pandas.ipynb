```python
 from time import process_time
```

```python
list1 = [i for i in range(10000)]
starttime = process_time()
list1 = [i+5 for i in list1]
endtime = process_time()
print(endtime - starttime)
print(list1)
```

```python
np1 = np.array([i for i in range(10000)])
starttime = process_time()
np1 += 5
endtime = process_time()
print(endtime - starttime)
```

Bar plot

0,0 means origin and 1,1 rep height and width of rect

```python
fig=plt.figure()
ax= fig.add_axes([0,0,1,1])
language =['english',"ger","spanish","french","italian"]
people =[100,200,300,400,500]
ax.bar(language,people)
plt.show()
```

```python
fig=plt.figure()
ax= fig.add_axes([0,0,1,1])
```

PIE CHART

```python
fig1=plt.figure()
ax= fig1.add_axes([0,0,1,1])
language =['english',"ger","spanish","french","italian"]
people =[100,200,300,400,500]
ax.pie(people, labels = language, autopct='%1.1f%%')
plt.show()
```

scatterplot

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
plt.scatter(x,y, color ='g')
plt.scatter(x,z, color ='r')
plt.show()
```

3D scatterplot

```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.random(100) #100 random values b/w 0 and 1
x=np.sin(z)
y=np.cos(z)
ax.scatter3D(x,y,z,c=z, cmap='Greens')
plt.show()
```
